
# **Task 2: Data Cleaning**

**1. Replace "-" with NaN in the entire DataFrame**

In our dataset, missing values are represented as underscores. We'll replace these underscores with NaN (Not-a-Number) values using the replace function to facilitate further data processing.




---



```python
df.drop(columns=["Unnamed: 0", "Owner's Spouse Name"], inplace=True)
print('Total columns after drop:', len(df.columns))
df.head()
```

**2. Handling Missing Values:**

We identify and remove two columns, 'Unnamed: 0' and 'Owner's Spouse Name,' as they are not relevant to our analysis. Dropping unnecessary columns helps simplify the dataset.


---


```python
# Data Type Conversion

df['No of Resident'] = pd.to_numeric(df['No of Resident'], errors='coerce')  # Convert to numeric, converting errors to NaN
print(f"Data type of 'No of Resident' column : {df['No of Resident'].dtypes}")

df['Confirmed Members'] = pd.to_numeric(df['Confirmed Members'], errors='coerce')  # Convert to numeric, converting errors to NaN
print(f"Data type of 'Confirmed Members'column : {df['Confirmed Members'].dtypes}")

df['Maintenance Amt'] = pd.to_numeric(df['Maintenance Amt'], errors='coerce')  # Convert to numeric, converting errors to NaN
print(f"Data type of 'Maintenance Amt' column : {df['Maintenance Amt'].dtypes}")
```

**3.Data Type Conversion:**

We ensure that relevant columns are of the correct data types for analysis. Specifically, we convert 'No of Resident' and 'Confirmed Members' columns to numeric data types while handling any conversion errors by setting them to NaN.


---




```python
# Creating avg_maintenance_amt variable
avg_maintenance_amt = df.groupby('Flat Area (sq.mt)')['Maintenance Amt'].median().to_dict()

# Iterate rows to replace missing values in 'Maintenance Amt'
for index, row in df.iterrows():
    if pd.isnull(row['Maintenance Amt']):
        flat_area = row['Flat Area (sq.mt)']
        if flat_area in avg_maintenance_amt:
            df.at[index, 'Maintenance Amt'] = avg_maintenance_amt[flat_area]


# Replace missing values in 'Flat Vacancy' column based on 'Availability of owner'
df['Flat Vacancy'] = df['Flat Vacancy'].fillna(
    df['Availlability of owner'].apply(
        lambda x: 'Vacant' if x != 'Yes' else 'Owned'
    )
)
```


**4.Impute Missing Values in 'Maintenance Amt' , Flat Vacancy, 'Donation', No of Resident' , 'Confirmed Members', 'Availability of owner' and 'Origin of Owner'**

* To address missing values in the 'Maintenance Amt' column, we iteratively go through the DataFrame. For each missing value, we impute it based on the corresponding 'Flat Area (sq.mt)' value. We reference a previously calculated dictionary, 'avg_maintenance_amt,' to fill in these values.

* We handle missing values in the 'Flat Vacancy' column by using a conditional approach. If 'Availability of owner' is 'Yes,' we assume the flat is owned ('Flat Vacancy' is 'Owned'), and if 'Availability of owner' is not 'Yes,' we consider the flat as vacant ('Flat Vacancy' is 'Vacant').



---




```python
# Replace missing values in 'Donation', No of Resident' , 'Confirmed Members' with median
cols = ['Donation', 'No of Resident' , 'Confirmed Members']

for col in cols:
    df[col] = df[col].fillna(df[col].median())

# Replace missing values in 'Availability of owner', 'Origin of Owner' with mode
cols = ['Availlability of owner', 'Origin of Owner']

for col in cols:
    df[col] = df[col].fillna(df[col].mode()[0])
df.isnull().sum()
```



* For missing values in the 'Donation', No of Resident' , 'Confirmed Members' column, we impute them with the median value of the respective column.
* Finally, we handle missing values in the 'Availability of owner' and 'Origin of Owner' columns by imputing them with the mode (most frequent value) of their respective columns.


---






```python
# Show cleaned data summary
data_type = pd.DataFrame(df.dtypes, columns=['Data Type'])
missing = pd.DataFrame(df.isnull().sum(), columns=['Missing Values'])
missing_percentage = pd.DataFrame(round(df.isnull().mean() * 100,2), columns=['Missing Percentage'])
unique = pd.DataFrame(df.nunique(), columns=['Unique Values'])
sample = pd.DataFrame(df.head().T)

summary = pd.concat([data_type, missing, missing_percentage, unique, sample], axis=1)
summary.rename(columns={0: 'Row_1',
                        1: 'Row_2',
                        2: 'Row_3',
                        3: 'Row_4',
                        4: 'Row_5'
                       }, inplace=True)
summary
```

 **1. Resident Participation:**

**2. Donations Analysis:**

Calculate the average donation amount by finding the mean of the 'Donation' column.
Calculate the total donation amount by summing the 'Donation' column.


---



```python
avg_donation = df['Donation'].mean()
total_donation = df ['Donation'].sum()

print('Average donation amount per flat is:', avg_donation)
print('Toatal donation amount is:', total_donation)
```

**3. Event Planning Recommendations:**


* Create a list called 'recommendations' to store our event planning suggestions based on the data analysis.

* Check if the average donation amount is greater than or equal to 5000, suggesting a higher budget for the event.

* Check if the mean percentage of participation is above 70%, indicating a good turnout expectation.

* Check if the total donation amount collected exceeds 50000, indicating sufficient funds.



---





```python
# Create a list called 'recommendations' to store our event planning suggestions based on the data analysis.
recommendations = []

# Check if the average donation amount is greater than or equal to 5000, suggesting a higher budget for the event.
if avg_donation>= 5000:
  recommendations.append("Consider a higher budget for the event.")
else:
  recommendations.append("Budget allocation may be optimized based on donations.")

# Check if the mean percentage of participation is above 70%, indicating a good turnout expectation
avg_participation_percentage = round(participation_percentage.mean(), 2)
if avg_participation_percentage > 70:
    recommendations.append("Expect a good turnout for the event.")
else:
    recommendations.append("Engage in more outreach to boost participation.")

# Check if the total donation amount collected exceeds 50000, indicating sufficient funds.
if total_donation > 50000:
  recommendations.append("Sufficient funds are available for event planning.")
else:
  recommendations.append("Consider additional fundraising efforts.")


# Print the calculated statistics and recommendations
print("Percentage of Residents' Participation in the Event:", avg_participation_percentage)
print("Average Donation Amount:", avg_donation)
print("Total Donation Amount Collected:", total_donation)
print("\nEvent Planning Recommendations:")
for recommendation in recommendations:
    print("-" + recommendation)
```
